```python
import time

def measure_llm(llm, prompt):
    start = time.time()
    llm.invoke(prompt)
    print("LLM Time:", round(time.time() - start, 3))


def measure_retrieval(retriever, query):
    start = time.time()
    retriever.get_relevant_documents(query)
    print("Retrieval Time:", round(time.time() - start, 3))


def measure_total(func, *args):
    start = time.time()
    func(*args)
    print("Total Time:", round(time.time() - start, 3))
```
